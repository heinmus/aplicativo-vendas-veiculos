```python
class Loja:
    def __init__(self, nome, endereco, telefone_whatsapp):
        self.nome = nome
        self.endereco = endereco
        self.telefone_whatsapp = telefone_whatsapp

    def __repr__(self):
        return f"Loja(Nome: {self.nome}, Endereço: {self.endereco}, WhatsApp: {self.telefone_whatsapp})"

# Example usage:
loja1 = Loja("Starcar Automóveis", "Rua das Flores, 123 - Centro, Cidade", "+5511987654321")
print(loja1)
```

```python
class Veiculo:
    def __init__(self, marca, modelo, versao, cor, quilometragem, tipo_cambio, tipo_combustivel, ano_fabricacao, ano_modelo, opcionais, preco, observacoes=None, fotos=None, vendido=False, loja=None):
        self.marca = marca
        self.modelo = modelo
        self.versao = versao
        self.cor = cor
        self.quilometragem = quilometragem # KM
        self.tipo_cambio = tipo_cambio # Ex: Manual, Automático
        self.tipo_combustivel = tipo_combustivel # Ex: Gasolina, Flex, Elétrico
        self.ano_fabricacao = ano_fabricacao
        self.ano_modelo = ano_modelo
        self.opcionais = opcionais if opcionais is not None else [] # Lista de opcionais (e.g., ['Ar Condicionado', 'Direção Hidráulica'])
        self.preco = preco
        self.observacoes = observacoes
        self.fotos = fotos if fotos is not None else [] # Lista de URLs ou caminhos para as fotos
        self.vendido = vendido # Boolean para indicar se foi vendido
        self.loja = loja # Objeto Loja ou ID da Loja associada

    def __repr__(self):
        return f"Veiculo(Marca: {self.marca}, Modelo: {self.modelo}, Ano: {self.ano_modelo}, Preço: R${self.preco:,.2f})"

# Example usage:
veiculo1 = Veiculo(
    marca="Fiat",
    modelo="Cronos",
    versao="1.3 Drive",
    cor="Branco",
    quilometragem=25000,
    tipo_cambio="Automático",
    tipo_combustivel="Flex",
    ano_fabricacao=2022,
    ano_modelo=2023,
    opcionais=["Ar Condicionado", "Direção Elétrica", "Vidros Elétricos", "Multimídia"],
    preco=85000.00,
    observacoes="Único dono, todas as revisões em dia.",
    fotos=["url_foto1.jpg", "url_foto2.jpg"],
    loja=loja1 # Associando o veículo à loja criada anteriormente
)
print(veiculo1)
print(f"Loja do veículo: {veiculo1.loja.nome}")
```

```python
class Usuario:
    def __init__(self, login, senha, tipo_usuario='vendedor', nome=None, comissao_total=0.0, gastos_registrados=None):
        self.login = login
        self.senha = senha # Em um sistema real, a senha seria hashada e não armazenada em texto plano
        self.tipo_usuario = tipo_usuario # Ex: 'admin', 'vendedor'
        self.nome = nome
        self.comissao_total = comissao_total
        self.gastos_registrados = gastos_registrados if gastos_registrados is not None else []

    def __repr__(self):
        return f"Usuario(Login: {self.login}, Tipo: {self.tipo_usuario}, Nome: {self.nome}, Comissão: R${self.comissao_total:,.2f}, Gastos: {len(self.gastos_registrados)})" # Simplified view of expenses count

    def registrar_comissao(self, valor_comissao):
        if valor_comissao > 0:
            self.comissao_total += valor_comissao
            print(f"Comissão de R${valor_comissao:,.2f} registrada para {self.nome}. Total: R${self.comissao_total:,.2f}")
        else:
            print("Valor de comissão deve ser positivo.")

    def registrar_gasto(self, descricao, valor):
        if valor > 0:
            self.gastos_registrados.append({'descricao': descricao, 'valor': valor})
            print(f"Gasto de '{descricao}' (R${valor:,.2f}) registrado para {self.nome}.")
        else:
            print("Valor do gasto deve ser positivo.")

    def listar_gastos(self):
        if not self.gastos_registrados:
            print(f"Nenhum gasto registrado para {self.nome}.")
            return
        print(f"\n--- Gastos de {self.nome} ---")
        for i, gasto in enumerate(self.gastos_registrados):
            print(f"{i+1}. {gasto['descricao']}: R${gasto['valor']:,.2f}")
        print("---------------------------")

# Example usage:
admin_user = Usuario("admin", "admin_temp", "admin", "Administrador") # Senha inicial para admin
vendedor1 = Usuario("joao.vendedor", "senha123", "vendedor", "João Silva")

print(admin_user)
print(vendedor1)

# Note sobre segurança: Em um ambiente de produção, senhas nunca devem ser armazenadas em texto simples.
# Elas devem ser hashadas (e.g., usando bcrypt) e verificadas de forma segura.

# Demonstrating new features
vendedor1.registrar_comissao(1500.00)
vendedor1.registrar_gasto("Gasolina", 200.50)
vendedor1.registrar_gasto("Almoço cliente", 50.00)
print(vendedor1)
vendedor1.listar_gastos()
```

```python
from collections import defaultdict

def gerar_relatorio_gastos_por_categoria(usuario):
    if not usuario.gastos_registrados:
        print(f"Nenhum gasto registrado para {usuario.nome}.")
        return

    gastos_por_categoria = defaultdict(float)
    for gasto in usuario.gastos_registrados:
        categoria = gasto['descricao'] # Usando a descrição como categoria
        valor = gasto['valor']
        gastos_por_categoria[categoria] += valor

    print(f"\n--- Relatório de Gastos por Categoria para {usuario.nome} ---")
    for categoria, total in gastos_por_categoria.items():
        print(f"- {categoria}: R${total:,.2f}")
    print(f"Total Geral: R${sum(gastos_por_categoria.values()):,.2f}")
    print("---------------------------------------------------")

# Example Usage:
# Assuming 'vendedor1' is already defined and has registered expenses
gerar_relatorio_gastos_por_categoria(vendedor1)

# Example with a user without expenses
admin_user_sem_gastos = Usuario("admin_gastos", "senha", "admin", "Admin Sem Gastos")
gerar_relatorio_gastos_por_categoria(admin_user_sem_gastos)
```

```python
# In-memory 'database' for our objects
lojas = []
veiculos = []
usuarios = []

# Adding our example objects to the lists for demonstration
lojas.append(loja1)
veiculos.append(veiculo1)
usuarios.append(admin_user)
usuarios.append(vendedor1)

print(f"Initial Lojas: {len(lojas)}")
print(f"Initial Veiculos: {len(veiculos)}")
print(f"Initial Usuarios: {len(usuarios)}")
```

```python
def cadastrar_loja(nome, endereco, telefone_whatsapp):
    nova_loja = Loja(nome, endereco, telefone_whatsapp)
    lojas.append(nova_loja)
    print(f"Loja '{nome}' cadastrada com sucesso!")
    return nova_loja

def listar_lojas():
    if not lojas:
        print("Nenhuma loja cadastrada.")
        return []
    print("\n--- Lojas Cadastradas ---")
    for i, loja in enumerate(lojas):
        print(f"{i+1}. {loja}")
    print("-------------------------")
    return lojas

# Example Usage:
cadastrar_loja("Carros Premium", "Av. Principal, 500", "+5511998877665")
listar_lojas()
```

```python
def cadastrar_veiculo(marca, modelo, versao, cor, quilometragem, tipo_cambio, tipo_combustivel, ano_fabricacao, ano_modelo, opcionais, preco, loja, observacoes=None, fotos=None):
    novo_veiculo = Veiculo(marca, modelo, versao, cor, quilometragem, tipo_cambio, tipo_combustivel, ano_fabricacao, ano_modelo, opcionais, preco, observacoes, fotos, loja=loja)
    veiculos.append(novo_veiculo)
    print(f"Veículo '{marca} {modelo}' cadastrado com sucesso para a loja '{loja.nome}'!")
    return novo_veiculo

def listar_veiculos():
    if not veiculos:
        print("Nenhum veículo cadastrado.")
        return []
    print("\n--- Veículos Cadastrados ---")
    for i, veiculo in enumerate(veiculos):
        print(f"{i+1}. {veiculo} - Loja: {veiculo.loja.nome}")
    print("----------------------------")
    return veiculos

# Example Usage:
loja_atual = lojas[0] # Using the first store from our list
if len(lojas) > 1: # If we have more than one store, use the second one for variety
    loja_para_novo_veiculo = lojas[1]
else:
    loja_para_novo_veiculo = loja_atual # Otherwise, use the existing one

cadastrar_veiculo(
    marca="Chevrolet",
    modelo="Onix",
    versao="1.0 LT",
    cor="Preto",
    quilometragem=10000,
    tipo_cambio="Manual",
    tipo_combustivel="Flex",
    ano_fabricacao=2021,
    ano_modelo=2022,
    opcionais=["Ar Condicionado", "Direção Elétrica"],
    preco=72000.00,
    loja=loja_para_novo_veiculo,
    observacoes="Revisões em concessionária."
)
listar_veiculos()
```

```python
def cadastrar_usuario(login, senha, tipo_usuario='vendedor', nome=None):
    novo_usuario = Usuario(login, senha, tipo_usuario, nome)
    usuarios.append(novo_usuario)
    print(f"Usuário '{login}' ({tipo_usuario}) cadastrado com sucesso!")
    return novo_usuario

def listar_usuarios():
    if not usuarios:
        print("Nenhum usuário cadastrado.")
        return []
    print("\n--- Usuários Cadastrados ---")
    for i, user in enumerate(usuarios):
        print(f"{i+1}. {user}")
    print("---------------------------")
    return usuarios

# Example Usage:
cadastrar_usuario("maria.vendedora", "vendas456", "vendedor", "Maria Antônia")
listar_usuarios()
```

```python
def marcar_veiculo_vendido(veiculo_a_marcar):
    for veiculo in veiculos:
        if veiculo == veiculo_a_marcar:
            veiculo.vendido = True
            print(f"Veículo '{veiculo.marca} {veiculo.modelo}' marcado como vendido.")
            return True
    print("Veículo não encontrado.")
    return False

def remover_veiculo(veiculo_a_remover):
    global veiculos
    initial_len = len(veiculos)
    veiculos = [v for v in veiculos if v != veiculo_a_remover]
    if len(veiculos) < initial_len:
        print(f"Veículo '{veiculo_a_remover.marca} {veiculo_a_remover.modelo}' removido do catálogo.")
        return True
    print("Veículo não encontrado para remoção.")
    return False

# Example Usage:
print("\n--- Antes de Marcar/Remover ---")
listar_veiculos()

# Mark veiculo1 as sold
marcar_veiculo_vendido(veiculo1)

print("\n--- Após Marcar Veículo 1 como Vendido ---")
listar_veiculos()

# Let's add another vehicle to demonstrate removal later
cadastrar_veiculo(
    marca="Ford",
    modelo="Ka",
    versao="SE 1.0",
    cor="Prata",
    quilometragem=50000,
    tipo_cambio="Manual",
    tipo_combustivel="Flex",
    ano_fabricacao=2019,
    ano_modelo=2020,
    opcionais=["Ar Condicionado", "Direção Hidráulica"],
    preco=45000.00,
    loja=lojas[0] # Associating with the first store
)

# Get the newly added vehicle to remove it
veiculo_a_re_remover = veiculos[-1]

print("\n--- Antes de Remover Veículo (Ford Ka) ---")
listar_veiculos()

# Remove the last added vehicle (Ford Ka)
remover_veiculo(veiculo_a_re_remover)

print("\n--- Após Remover Veículo (Ford Ka) ---")
listar_veiculos()
```

```python
def buscar_veiculos(termo_busca, campo='modelo'):
    resultados = []
    termo_busca_lower = termo_busca.lower()
    for veiculo in veiculos:
        if campo == 'modelo' and termo_busca_lower in veiculo.modelo.lower():
            resultados.append(veiculo)
        elif campo == 'marca' and termo_busca_lower in veiculo.marca.lower():
            resultados.append(veiculo)
        # You can expand this to search other fields like 'versao', 'cor', etc.
        # elif campo == 'cor' and termo_busca_lower in veiculo.cor.lower():
        #     resultados.append(veiculo)
    return resultados

# Example Usage:
print("\n--- Busca por Veículos ---")

# Adding a few more vehicles to make search more interesting
cadastrar_veiculo("Hyundai", "HB20", "Comfort Plus", "Prata", 30000, "Automático", "Flex", 2020, 2021, ["Ar Condicionado", "Multimídia"], 65000.00, lojas[1])
cadastrar_veiculo("Fiat", "Palio", "Attractive", "Vermelho", 70000, "Manual", "Flex", 2016, 2016, ["Ar Condicionado"], 35000.00, lojas[0])

print("\n--- Todos os veículos para referência ---")
listar_veiculos()

print("\n--- Buscando por 'Cronos' (modelo) ---")
resultados_cronos = buscar_veiculos("cronos", campo='modelo')
if resultados_cronos:
    for v in resultados_cronos:
        print(v)
else:
    print("Nenhum veículo encontrado para 'Cronos'.")

print("\n--- Buscando por 'Fiat' (marca) ---")
resultados_fiat = buscar_veiculos("fiat", campo='marca')
if resultados_fiat:
    for v in resultados_fiat:
        print(v)
else:
    print("Nenhum veículo encontrado para 'Fiat'.")

print("\n--- Buscando por 'Civic' (modelo inexistente) ---")
resultados_civic = buscar_veiculos("civic", campo='modelo')
if resultados_civic:
    for v in resultados_civic:
        print(v)
else:
    print("Nenhum veículo encontrado para 'Civic'.")
```
